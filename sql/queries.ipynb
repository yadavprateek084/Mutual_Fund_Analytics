```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../database/mutual_fund.db")
```

```python
query = """
SELECT
    amfi_code,
    aum_crore
FROM fact_performance
ORDER BY aum_crore DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
query = """

-- QUERY 1: Top 5 Funds by AUM


SELECT
    amfi_code,
    aum_crore
FROM fact_performance
ORDER BY aum_crore DESC
LIMIT 5;



"""

pd.read_sql(query, conn)
```

```python
query = """


-- =====================================================
-- QUERY 3: Total SIP Inflow by Year
-- =====================================================

SELECT
    strftime('%Y', month) AS year,
    SUM(sip_inflow_crore) AS total_sip_inflow
FROM fact_sip
GROUP BY year
ORDER BY year;



"""

pd.read_sql(query, conn)
```

```python
query = """


-- =====================================================
-- QUERY 4: Transactions by State
-- =====================================================

SELECT
    state,
    COUNT(*) AS total_transactions,
    SUM(amount_inr) AS total_amount
FROM fact_transactions
GROUP BY state
ORDER BY total_amount DESC;



"""

pd.read_sql(query, conn)
```

```python
query = """



-- =====================================================
-- QUERY 5: Funds with Expense Ratio Below 1%
-- =====================================================

SELECT
    scheme_name,
    fund_house,
    expense_ratio_pct
FROM dim_fund
WHERE expense_ratio_pct < 1
ORDER BY expense_ratio_pct;



"""

pd.read_sql(query, conn)
```

```python
query = """



-- =====================================================
-- QUERY 6: Top 10 Fund Houses by AUM
-- =====================================================

SELECT
    fund_house,
    MAX(aum_crore) AS latest_aum
FROM fact_aum
GROUP BY fund_house
ORDER BY latest_aum DESC
LIMIT 10;





"""

pd.read_sql(query, conn)
```

```python
query = """


-- =====================================================
-- QUERY 7: Average Transaction Amount by Type
-- =====================================================

SELECT
    transaction_type,
    ROUND(AVG(amount_inr),2) AS avg_amount
FROM fact_transactions
GROUP BY transaction_type;



"""

pd.read_sql(query, conn)
```

```python
query = """




-- =====================================================
-- QUERY 8: Top Performing Funds by Sharpe Ratio
-- =====================================================

SELECT
    d.scheme_name,
    p.sharpe_ratio
FROM fact_performance p
JOIN dim_fund d
ON p.amfi_code = d.amfi_code
ORDER BY p.sharpe_ratio DESC
LIMIT 10;



"""

pd.read_sql(query, conn)
```

```python
query = """



-- =====================================================
-- QUERY 9: Investor Distribution by Age Group
-- =====================================================

SELECT
    age_group,
    COUNT(DISTINCT investor_id) AS investors
FROM fact_transactions
GROUP BY age_group
ORDER BY investors DESC;



"""

pd.read_sql(query, conn)
```

```python
query = """



-- =====================================================
-- QUERY 10: Benchmark Index Performance
-- =====================================================

SELECT
    index_name,
    ROUND(MIN(close_value),2) AS min_value,
    ROUND(MAX(close_value),2) AS max_value,
    ROUND(AVG(close_value),2) AS avg_value
FROM fact_benchmark
GROUP BY index_name
ORDER BY avg_value DESC;


"""

pd.read_sql(query, conn)
```

```python
query = """

-- Top 5 States by Investment Amount

SELECT
    state,
    SUM(amount_inr) AS total_investment
FROM fact_transactions
GROUP BY state
ORDER BY total_investment DESC
LIMIT 5;

"""

pd.read_sql(query, conn)
```

```python
query = """

-- SIP vs Lumpsum vs Redemption

SELECT
    transaction_type,
    COUNT(*) AS transaction_count,
    SUM(amount_inr) AS total_amount
FROM fact_transactions
GROUP BY transaction_type;

"""

pd.read_sql(query, conn)
```
